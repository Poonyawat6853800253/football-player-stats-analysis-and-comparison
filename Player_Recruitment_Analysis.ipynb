### Player Recruitment Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("players_data_light-2024_2025.csv")

print(df.shape)
df.head()
```

```python
recruit = df.copy()

# กรองคนที่ลงสนามอย่างน้อย 900 นาที
recruit = recruit[recruit["Min"] >= 900].copy()

print(recruit.shape)
recruit.head()
```

```python
recruit["Gls_per90"] = recruit["Gls"] / recruit["90s"]
recruit["Ast_per90"] = recruit["Ast"] / recruit["90s"]
recruit["GA_per90"] = recruit["G+A"] / recruit["90s"]
recruit["xG_per90"] = recruit["xG"] / recruit["90s"]
recruit["xAG_per90"] = recruit["xAG"] / recruit["90s"]
recruit["PrgC_per90"] = recruit["PrgC"] / recruit["90s"]
recruit["PrgP_per90"] = recruit["PrgP"] / recruit["90s"]
recruit["PrgR_per90"] = recruit["PrgR"] / recruit["90s"]

recruit[[
    "Player", "Squad", "Comp", "Pos", "Age", "Min",
    "Gls_per90", "Ast_per90", "GA_per90"
]].head()
```

```python
forwards = recruit[recruit["Pos"].str.contains("FW", na=False)].copy()

target_fw = forwards[
    (forwards["Age"] <= 25) &
    (forwards["Min"] >= 900)
].copy()

target_fw = target_fw.sort_values("GA_per90", ascending=False)

target_fw[[
    "Player", "Squad", "Comp", "Age", "Min",
    "Gls", "Ast", "GA_per90", "xG_per90", "xAG_per90"
]].head(20)
```

```python
features = [
    "Gls_per90",
    "Ast_per90",
    "xG_per90",
    "xAG_per90",
    "PrgC_per90",
    "PrgP_per90",
    "PrgR_per90"
]

target_fw = target_fw.dropna(subset=features).copy()

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(target_fw[features])

target_fw["Recruitment_Score"] = scaled_features.mean(axis=1)

target_fw = target_fw.sort_values("Recruitment_Score", ascending=False)

target_fw[[
    "Player", "Squad", "Comp", "Age", "Min",
    "Gls", "Ast", "Recruitment_Score"
]].head(20)
```

```python
weights = {
    "Gls_per90": 0.30,
    "xG_per90": 0.25,
    "Ast_per90": 0.10,
    "xAG_per90": 0.10,
    "PrgC_per90": 0.10,
    "PrgP_per90": 0.10,
    "PrgR_per90": 0.05
}

scaled_df = pd.DataFrame(
    scaled_features,
    columns=features,
    index=target_fw.index
)

target_fw["Weighted_Recruitment_Score"] = sum(
    scaled_df[col] * weight for col, weight in weights.items()
)

target_fw = target_fw.sort_values("Weighted_Recruitment_Score", ascending=False)

target_fw[[
    "Player", "Squad", "Comp", "Age", "Min",
    "Gls", "Ast", "Weighted_Recruitment_Score"
]].head(20)
```
